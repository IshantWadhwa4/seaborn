**Seaborn**

```
import seaborn as sns
```

Now first we will start with uni varient data.
i.e for single collumn.
in Seaborn for univariant data there are maps like 
distributin plot(distplot) 

```
sns.distplot(wine_data.alcohol)
# area under the curve show the problistic distribution or KDE it is almost normal distributed
```

```
sns.distplot(wine_data.alcohol,kde=False)
# without KDE its a hist graph
```

```
sns.distplot(wine_data.alcohol,bins=80,rug=True,kde=True)
# rug give the line on x-axis with exact point value
```

```
sns.distplot(wine_data.alcohol,bins=80,rug=True,kde=True,hist=False)
```

```
import matplotlib.pyplot as plt
```

```
f,ax = plt.subplots(figsize=(10,5)) # seaborn is based on matplot lib sor figure and axis size

sns.set(color_codes=True) # add color  
sns.kdeplot(wine_data.alcohol,shade=True,color= 'blue')
# this is KDE plot 
```

**Bi Variant**

For two data columns
Join plots

```
sns.jointplot(x="quality",y="alcohol",data=wine_data)
# show relation b/w alcohol and quality we have one numeric contious and one categorical
```

```
sns.jointplot(x="free sulfur dioxide",y="total sulfur dioxide",data=wine_data)
```

```
with sns.axes_style('white'):
    sns.jointplot(x="free sulfur dioxide",y="total sulfur dioxide",data=wine_data,
                 kind='hex',color='k',xlim={0,100},ylim={0,250})
# add value to each point in hex form    
```

```
with sns.axes_style('white'):
    sns.jointplot(x="free sulfur dioxide",y="total sulfur dioxide",data=wine_data,
                 kind='kde',color='k',xlim={0,100},ylim={0,250})
```

Relation ship betwwn more than one Heatmap

```
cor = wine_data.corr()
f, ax=plt.subplots(figsize=(12, 12))
sns.heatmap(cor,vmax=.9,annot=True,fmt='.2f',cmap="summer")
```
